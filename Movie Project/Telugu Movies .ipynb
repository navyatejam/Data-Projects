```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("~/Downloads/Movies 1940-1969.csv")
```

```python
data.head()
```

```python
data.columns = ['Movie', 'Director', 'Actors']
```

```python
## Writing each column as a key and value
data.to_dict()
```

```python
data.to_dict('records')
```

```python
## Writing each row as a key and value

Dict = data.to_dict('records')
print(Dict)
```

## Given a movie name, we can find the actors


```python
Movie_input = input('Enter your movie:')
for r in range(len(Dict)):
    if Dict[r]["Movie"] == Movie_input:
        print("The actors in {} are {}." .format(Dict[r]['Movie'], Dict[r]['Actors']))
   
   
```

## Given a movie name, we can find the actors and the director of the movie

```python
Movie_input = input('Enter your movie:')
for r in range(len(Dict)):
    if Dict[r]["Movie"] == Movie_input:
        print("The actors in {} are {} and the director of the movie {} is {}." .format(Dict[r]['Movie'], Dict[r]['Actors'], Dict[r]['Movie'], Dict[r]['Director']))
   
   
```

## Listing all the movies directed by the given director
